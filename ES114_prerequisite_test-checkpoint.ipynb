```python
#1.List Operations and Basic Statistics:
import math
nums=[]
term=1

#creating list of the first 20 natural numbers
while term<=20:
    nums.append(term)
    term+=1

#computing sum of all even numbers in the list and computing product of all odd numbers in the list
summation=0
product=1
for i in nums:
    if i%2==0:
        summation+=i
    else:
        product*=i


#computing mean,median and standard deviation of the list
def mean_of_list(l):
    sum_of_terms=0
    for i in l:
        sum_of_terms+=i
    mean=(sum_of_terms)/len(l)
    return mean

def median_of_list(l):
    l.sort()
    if len(l)%2!=0:
        target=(len(l)+1)/2
        return l[int(target)]
    else:
        value=(l[int(len(l)/2-1)]+l[int(len(l)/2)])/2 #taking average of 11th and 12th term instead of 10th and 11th because list indexing starts from 0
        return value
def standard_deviation_of_list(l):
    mean=mean_of_list(l)
    sum_of_all_difference_squares=0
    for i in l:
        sum_of_all_difference_squares+=(mean-i)**2
    variance=sum_of_all_difference_squares/len(l)
    return math.sqrt(variance)

#displaying all the values
print("sum of all even numbers:",summation)
print("product of all odd numbers:",product)
print("mean:",mean_of_list(nums))
print("median:",median_of_list(nums))
print("standard deviation:",standard_deviation_of_list(nums))
    
    
        
    
```

2.Mutable vs. Immutable Data Types:
Mutable Data Types: These are data types that can have their value changed after defining. Example of Mutable data types are lists and dictionaries.
Immutable Data types: These are data types whose value cannot be changed once defined. Example of Immutable data types are strings and tuples.

```python
#3.Fibonacci Sequence:
import matplotlib.pyplot as plt
n=int(input())

#defining two lists,one which stores the value of x and one which stores f(x), f(x) here is the fibonacci value at x.
x_val=[0,1]
fibo_val=[0,1]

#calculating and storing values of f(x) at each value of x up till n.
for i in range(2,n):
    x_val.append(i)
    term=fibo_val[i-1]+fibo_val[i-2]
    fibo_val.append(term)

#plotting x values against values of f(x)
plt.title("Line Graph")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x_val,fibo_val)
plt.show()
```

```python
#15.Anagram Finder:
word=input("Enter Word: ")
words=input("Enter space seperated words").split()

#function that checks if two words are anagrams
def anagram(x,y):
    x_list=[]
    y_list=[]
    for i in x:
        x_list.append(i)
    for i in y:
        y_list.append(i)
    x_list.sort()
    y_list.sort()
    for i in range(len(x_list)):
        if x_list[i]==y_list[i]:
            continue
        else:
            return False
    return True

#storing and displaying all anagrams
anagrams=[]
for term in words:
    if anagram(word,term):
        anagrams.append(term)
print("Anagrams of",word,"are:",anagrams)
```

```python
#16.Set Operations:
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Function to compute and visualize list operations
def list_operations_visualization(list1, list2):
    # Convert lists to sets for operations
    set1 = set(list1)
    set2 = set(list2)
    
    # Compute intersection, union, and symmetric difference
    intersection = set1 & set2
    union = set1 | set2
    symmetric_difference = set1 ^ set2
    
    # Print results
    print(f"List 1: {list1}")
    print(f"List 2: {list2}")
    print(f"Intersection: {list(intersection)}")
    print(f"Union: {list(union)}")
    print(f"Symmetric Difference: {list(symmetric_difference)}")
    
    # Visualize using a Venn diagram
    plt.figure(figsize=(8, 6))
    venn = venn2([set1, set2], ('List 1', 'List 2'))
    venn.get_label_by_id('10').set_text(f"Only in List 1\n{set1 - set2}")
    venn.get_label_by_id('01').set_text(f"Only in List 2\n{set2 - set1}")
    venn.get_label_by_id('11').set_text(f"Intersection\n{intersection}")
    plt.title("Venn Diagram of List Operations")
    plt.show()

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
list_operations_visualization(list1, list2)
```

```python
#20.CSV File Operations:
import csv
# Function to read a CSV file and compute the sum of values in a specific column
def read_csv_and_compute_sum(file_path, column_name):
    try:
        with open(file_path, mode='r') as file:
            csv_reader = csv.DictReader(file)  # Read as a dictionary for easy access by column name
            total = 0
            values = []
            
            for row in csv_reader:
                if column_name in row:
                    # Accumulate the sum (assuming numeric data in the column)
                    value = float(row[column_name])
                    total += value
                    values.append(value)
                else:
                    print(f"Column '{column_name}' not found in the file.")
                    return
            
            print(f"The sum of the values in the column '{column_name}' is: {total}")
            
            # Visualize the data
            plot_data(values, column_name)
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Function to visualize the data as a line plot
def plot_data(values, column_name):
    import matplotlib.pyplot as plt

    plt.figure(figsize=(10, 6))
    plt.plot(range(len(values)), values, marker='o', label=column_name)
    plt.title(f"Line Plot of {column_name}")
    plt.xlabel("Index")
    plt.ylabel(column_name)
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
file_path = "example.csv"  # Replace with the path to your CSV file
column_name = "Values"    # Replace with the column name you wan
read_csv_and_compute_sum(file_path, column_name)
```

```python
#21.Basic Statistics:
import math
nums = list(map(int,input("Enter a comma seperated list").split(","))) #taking input

#computing mean,median,mode and standard deviation of the list
def mean_of_list(l):
    sum_of_terms=0
    for i in l:
        sum_of_terms+=i
    mean=(sum_of_terms)/len(l)
    return mean

def median_of_list(l):
    l.sort()
    if len(l)%2!=0:
        target=(len(l)+1)/2
        return l[int(target)]
    else:
        value=(l[int(len(l)/2-1)]+l[int(len(l)/2)])/2 
        return value

def mode_of_list(l):
    val_freq={} #creating a dictionary that keeps count of the number of times a number appears in the list
    for i in l:
        if i not in val_freq:
            val_freq[i]=1
        else:
            val_freq[i]+=1
    maximum=0
    for i in val_freq:
        if val_freq[i]>maximum:
            maximum=val_freq[i]
    modes=[]
    for i in val_freq:
        if val_freq[i]==maximum:
            modes.append(i)
    result = ", ".join(map(str, modes))
    print(result)
        
def standard_deviation_of_list(l):
    mean=mean_of_list(l)
    sum_of_all_difference_squares=0
    for i in l:
        sum_of_all_difference_squares+=(mean-i)**2
    variance=sum_of_all_difference_squares/len(l)
    return math.sqrt(variance)

#displaying all calculated values
print("mean:",mean_of_list(nums))
print("median:",median_of_list(nums))
print("mode:",end=" ")
mode_of_list(nums)
print("standard deviation:",standard_deviation_of_list(nums))


#visualizing data using a box plot
import matplotlib.pyplot as plt
plt.title("Box Plot of data set")
plt.xlabel("Data")
plt.ylabel("Frequency distribution")
plt.boxplot(nums)
plt.show()
```

```python
#22.Normalization:
nums=list(map(int,input("Enter array of numbers to be normalized").split()))

#writing function that performs normalization within 0 and 1
def normalizer(array):
    normalized_values=[]
    array.sort()
    minimum=array[0]
    maximum=array[-1]
    for i in array:
        term= (((i-minimum)/(maximum-minimum))*(1)) + 0
        normalized_values.append(term)
    return normalized_values

#visualizing data using line plot
plt.title("Normalization")
plt.plot(nums,label="Original values")
plt.plot(normalizer(nums),label="Normalized values")
plt.legend()
plt.show
```

```python
#24.Grade Calculator:
mark=int(input("Enter score between 0 and 100:"))

#Creating a function that assignes grades based on mark 
def grade_finder(x):
    if x<=100 and x>90:
        return "A"
    elif x<=90 and x>80:
        return "B"
    elif x<=80 and x>70:
        return "C"
    elif x<=70 and x>60:
        return "D"
    else:
        return "F"

#Displaying the grade assigned
print("Grade assigned:",grade_finder(mark))
```

```python
#25.Leap Year Checker:
year=int(input("Enter year to be checked:"))

#Creating  a function to check if a year is a leap year with necessary constraints 
def leap_year_check(x):
    if x%4==0 and x%100!=0:
        return True
    elif x%100==0 and x%400==0:
        return True
    return False

#Displaying output
if leap_year_check(year):
    print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```
